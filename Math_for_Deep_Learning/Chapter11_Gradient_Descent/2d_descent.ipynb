```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def f(x,y):
    return 6*x**2 + 9*y**2 - 12*x - 14*y + 3

def dx(x):
    return 12*x - 12

def dy(y):
    return 18*y - 14
```

```python
N = 100
x,y = np.meshgrid(np.linspace(-1,3,N),np.linspace(-1,3,N))
z = f(x,y)
plt.contourf(x,y,z, 10, cmap = "jet")
plt.contour(x,y,z, 10, colors = "black", linewidths = 1)
plt.plot([0,0],[-1,3],color = 'k', linewidth = 1)
plt.plot([-1,3],[0,0],color = 'k', linewidth = 1)
plt.plot(1,0.77777778, color = 'k', marker = '+')

xold = xnew = -0.5
yold = ynew = 2.9
for i in range(24):
    plt.plot([xold,xnew], [yold,ynew], marker = 'o', color = 'k', linestyle = 'dotted')
    xold = xnew
    yold = ynew
    xnew = xold - 0.01*dx(xold)
    ynew = yold - 0.01*dy(yold)

xold = xnew = 2.9
yold = ynew = 2.9
for i in range(12):
    plt.plot([xold,xnew], [yold,ynew], marker = '<', color = 'k', linestyle = 'dotted')
    xold = xnew
    yold = ynew
    xnew = xold - 0.05*dx(xold)
    ynew = yold - 0.05*dy(yold)

xold = xnew = 2.9
yold = ynew = -0.7
for i in range(24):
    plt.plot([xold,xnew], [yold,ynew], marker = 's', color = 'k', linestyle = 'dotted')
    xold = xnew
    yold = ynew
    xnew = xold - 0.1*dx(xold)
    ynew = yold - 0.01*dy(yold)

plt.xlabel("$x$")
plt.ylabel("$y$")
plt.tight_layout(pad=0, w_pad=0, h_pad=0)
plt.show()
```
