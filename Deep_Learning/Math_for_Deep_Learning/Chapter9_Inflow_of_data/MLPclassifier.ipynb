```python
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(8675309)
x0 = np.random.random(50) - 0.3
y0 = np.random.random(50) + 0.3
x1 = np.random.random(50) + 0.3
y1 = np.random.random(50) - 0.3
print("x0,y0: %0.6f, %0.6f" % (x0.mean(), y0.mean()))
print("x1,y1: %0.6f, %0.6f" % (x1.mean(), y1.mean()))
print()

x = np.zeros((100, 2))
x[:50, 0] = x0; x[:50, 1] = y0
x[50:, 0] = x1; x[50:, 1] = y1
y = np.array([0] * 50 + [1] * 50)

idx = np.argsort(np.random.random(100))
x = x[idx]; y = y[idx]
x_train = x[:75]; x_test = x[75:]
y_train = y[:75]; y_test = y[75:]
```

```python
plt.plot(x0,y0,marker='o',linestyle='none')
plt.plot(x1,y1,marker='s',linestyle='none')
plt.xlabel(r'$x_0$')
plt.ylabel(r'$x_1$')
plt.tight_layout(pad=0, w_pad=0, h_pad=0)
plt.show()
```

```python
clf = MLPClassifier(hidden_layer_sizes = (5,), max_iter = 2000)
clf.fit(x_train, y_train)

print('Training accuracy: %.2f' % clf.score(x_train, y_train))
print('Test accuracy: %.2f' % clf.score(x_test, y_test))

W0 = clf.coefs_[0].T
b0 = clf.intercepts_[0].reshape((5,1))
W1 = clf.coefs_[1].T
b1 = clf.intercepts_[1].reshape((1,1))

print("Weights and biases:")
print(W0)
print(b0)
print()
print(W1)
print(b1)
print()
```
